```python
import tensorflow as tf
```

```python
file_name_queue = tf.train.string_input_producer(['../datas/data-01-test-score.csv'],
                                                shuffle=False, name='filename_queue')
```

```python
reader = tf.TextLineReader()
key, value = reader.read(file_name_queue)
```

```python
record_defaults = [[0.], [0.], [0.], [0.]]
xy = tf.decode_csv(value, record_defaults = record_defaults)
```

```python
train_x_batch, train_y_batch = tf.train.batch([xy[0:-1], xy[-1:]], batch_size = 10)
```

```python
# placeholder
X = tf.placeholder(tf.float32, shape = [None, 3])
Y = tf.placeholder(tf.float32, shape = [None, 1])
```

```python
# Variables
W = tf.Variable(tf.random_normal([3, 1]), name = 'weight')
b = tf.Variable(tf.random_normal([1]), name = 'bias')
```

```python
# Hypothesis
hypothesis = tf.matmul(X, W) + b
```

```python
# cost
cost = tf.reduce_mean(tf.square(hypothesis - Y))
```

```python
# Minimize
optimizer = tf.train.GradientDescentOptimizer(learning_rate = 1e-5)
train = optimizer.minimize(cost)
```

```python
# prepare session
session = tf.Session()
session.run(tf.global_variables_initializer())
```

```python
# Start populating the filename queue.
coord = tf.train.Coordinator()
threads = tf.train.start_queue_runners(sess=session, coord=coord)
```

```python
for step in range(10001):
    x_batch, y_batch = session.run([train_x_batch, train_y_batch])
    cost_val, hypothesis_val, _ = session.run([cost, hypothesis, train],
                                              feed_dict={X: x_batch, Y: y_batch})
    if step % 100 == 0:
        print("x_batch: ", x_batch, "\ny_batch: ", y_batch)
        print(step, "Cost: ", cost_val, "\nPrediction: ", hypothesis_val)
```

```python
coord.request_stop()
coord.join(threads)
```
