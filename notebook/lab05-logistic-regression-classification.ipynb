```python
import tensorflow as tf
```

```python
# train data
x_data = [[1,2], [2,3], [3,1], [4,3], [5,3], [6,2]]
y_data = [[0], [0], [0], [1], [1], [1]]
```

```python
# placeholder
X = tf.placeholder(tf.float32, shape = [None, 2])
Y = tf.placeholder(tf.float32, shape = [None, 1])
```

```python
# Variables
W = tf.Variable(tf.random_normal([2,1]), name='weight')
b = tf.Variable(tf.random_normal([1]), name='bias')
```

```python
# hypothesis using sigmoid: tf.div(1., 1 + tf.exp(tf.matmul(X, W) + b))
hypothesis = tf.sigmoid(tf.matmul(X, W) + b)
```

```python
# cost/loss function
cost = -tf.reduce_mean(Y * tf.log(hypothesis) + (1 - Y) * tf.log(1 - hypothesis))
```

```python
# minimize cost
optimizer = tf.train.GradientDescentOptimizer(learning_rate = 0.01)
train = optimizer.minimize(cost)
```

```python
# Accuraty computation
# True if hypothesis > 0.5 else False
# hypothesis > 0.5: True, False => tf.cast() -> 0 ~ 1사이 값이로 casting함.
predicted = tf.cast(hypothesis > 0.5, dtype = tf.float32)
accuracy = tf.reduce_mean(tf.cast(tf.equal(predicted, Y), dtype = tf.float32))
```

```python
# session
session = tf.Session()
session.run(tf.global_variables_initializer())
```

```python
# train the model
for step in range(10001):
    cost_val, h_val, _ = session.run([cost, hypothesis, train],
                                     feed_dict={X: x_data, Y: y_data})
    if step % 200 == 0:
        print(step, cost_val, h_val)

h_val, p_val, a_val = session.run([hypothesis, predicted, accuracy],
                                  feed_dict = {X: x_data, Y: y_data})
print("Hypothesis:", h_val, "\nPredicted:", p_val, "\nAccuracy:", a_val)
```
