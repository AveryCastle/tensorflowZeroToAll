```python
import tensorflow as tf
import numpy as np
```

```python
tf.set_random_seed(777)
```

```python
data = np.loadtxt('../datas/data-01-test-score.csv', delimiter = ',', dtype = np.float32)
```

```python
x_train = data[:, 0: -1]
y_train = data[:, [-1]]

print(x_train.shape, x_train, len(x_train))
print(y_train.shape, y_train, len(y_train))
```

```python
# placeholder
X = tf.placeholder(tf.float32, shape = [None, 3])
Y = tf.placeholder(tf.float32, shape = [None, 1])
```

```python
# Variables
W = tf.Variable(tf.random_normal([3, 1]), name = 'weight')
b = tf.Variable(tf.random_normal([1]), name = 'bias')
```

```python
# Hypothesis
hypothesis = tf.matmul(X, W) + b
```

```python
# Cost/Loss function
cost = tf.reduce_mean(tf.square(hypothesis - Y))
```

```python
# minimize cost
optimizer = tf.train.GradientDescentOptimizer(learning_rate = 1e-5)
train = optimizer.minimize(cost)
```

```python
# prepare session
session = tf.Session()
session.run(tf.global_variables_initializer())
```

```python
# fit
for step in range(5001):
    hypothesis_val, cost_val, _ = session.run([hypothesis, cost, train],
                                              feed_dict = {X: x_train, Y: y_train})
    if step % 200 == 0:
        print(step, "Cost: ", cost_val, "\n Prediction: ", hypothesis_val)
        
print("your score will be ", session.run(hypothesis,
                                         feed_dict={X: [[100, 90, 95]]}))
print("other score will be ", session.run(hypothesis,
                                          feed_dict={X: [[100, 90, 95], [60, 80, 90]]}))
```
