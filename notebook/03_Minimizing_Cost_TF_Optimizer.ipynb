# Minimizing Cost TF Optimizer

```python
import tensorflow as tf
```

### X and Y data

```python
X = [1, 2, 3]
Y = [1, 2, 3]
```

### Variables

```python
# Set the wrong model weight
W = tf.Variable(-4.)
```

### Model(Hypothesis)

```python
# hypothesis
hypothesis = W * X

# manual gradient
gradient = tf.reduce_mean((W * X - Y) * X) * 2

# cost/loss function
cost = tf.reduce_mean(tf.square(hypothesis - Y))
```

## Minimize cost

```python
optimizer = tf.train.GradientDescentOptimizer(learning_rate = 0.01)
train = optimizer.minimize(cost)
```

```python
# Get Gradient
gvs = optimizer.compute_gradients(cost, [W])

# Apply Gradient
apply_gradient = optimizer.apply_gradients(gvs)
```

## Prepare Session

```python
session = tf.Session()
session.run(tf.global_variables_initializer())
```

## Fit the line

```python
for step in range(2001):
    if step % 10 == 0:
        print(step, session.run([gradient, W, gvs]))
    # equal to session.run(train)
    session.run(apply_gradient)
```
