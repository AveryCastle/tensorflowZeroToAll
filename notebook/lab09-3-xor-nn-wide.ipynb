```python
import tensorflow as tf
import numpy as np
```

```python
tf.set_random_seed(777)
```

```python
x_data = [[0,0],
          [0,1],
          [1,0],
          [1,1]]
y_data = [[0],
          [1],
          [1],
          [0]]

x_data = np.array(x_data, dtype=np.float32)
y_data = np.array(y_data, dtype=np.float32)
```

```python
# Placeholder
X = tf.placeholder(tf.float32, shape=[None,2])
Y = tf.placeholder(tf.float32, shape=[None,1])
```

```python
# Variables - 1
W1 = tf.Variable(tf.random_normal([2,10]), name="weight1")
b1 = tf.Variable(tf.random_normal([10]), name="bias1")

layer = tf.sigmoid(tf.matmul(X, W1) + b1)
```

```python
# Variable - 2
W2 = tf.Variable(tf.random_normal([10, 1]), name='weight2')
b2 = tf.Variable(tf.random_normal([1]), name='bias2')
```

```python
# Hypothesis
hypothesis = tf.sigmoid(tf.matmul(layer, W2) + b2)
```

```python
# Cost
cost = -tf.reduce_mean(Y*tf.log(hypothesis) + (1-Y)*tf.log(1-hypothesis))
```

```python
# Minimize cost
train = tf.train.GradientDescentOptimizer(learning_rate=0.01).minimize(cost)
```

```python
# Accuracy
predicted = tf.cast(hypothesis > 0.5, dtype=tf.float32)
accuracy = tf.reduce_mean(tf.cast(tf.equal(predicted, Y), dtype=tf.float32))
```

```python
# Fit the train
with tf.Session() as session:
    session.run(tf.global_variables_initializer())
    
    for step in range(1, 10001):
        cost_val, hypothesis_val, _ = session.run([cost, hypothesis, train],
                                                  feed_dict={X: x_data, Y: y_data})
        if step % 100 == 0:
            print(step, hypothesis_val, cost_val)
    
    c_val, p_val, a_val = session.run([cost, predicted, accuracy],
                                      feed_dict={X: x_data, Y: y_data})
    print("C_val:", c_val, "\n, Expected:", p_val, ",\n Accuracy:", a_val)
```
